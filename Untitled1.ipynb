```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

class F:
    @staticmethod
    def sign(x):
        if x >= 0:
            return 1
        else:
            return -1

class Perceptron:
    def __init__(self, input_size):
        self.weights = np.random.randn(input_size)
        self.bias = np.random.randn()

    def forward(self, input):
        linear_output = np.dot(input, self.weights) + self.bias
        activation_output = F.sign(linear_output)
        return activation_output

class Optimizer:
    def __init__(self, model):
        self.model = model

    def update(self, x, y, y_hat):
        error = y - y_hat
        self.model.weights += error * x
        self.model.bias += error


X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=42)


for i in range(len(y)):
    if y[i] == 0:
        y[i] = -1


perceptron = Perceptron(input_size=2)
optimizer = Optimizer(perceptron)

n_iter = 1000
for _ in range(n_iter):
    for x_data, y_data in zip(X, y):
        y_hat = perceptron.forward(x_data)
        optimizer.update(x_data, y_data, y_hat)


plt.scatter(X[:, 0], X[:, 1], c=y)

x_line = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
y_line = -(perceptron.weights[0] * x_line + perceptron.bias) / perceptron.weights[1]
plt.plot(x_line, y_line, color='red')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data and Separator')
plt.show()
```
